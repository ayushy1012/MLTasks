Importing modules

```python
import pandas as pd
import numpy as np
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
df = pd.read_csv('/content/raw.csv',header=None)
```

First of all i converted the data in different columns (the csv side had everything in one column) converted the dataframe into float datatype and then checked range of all the columns and found that every column other than 0 and 1 had values distributed between 0.7 and 5 and hence column 0 and column 1 may be the label while other columns may be the features

```python
df.columns=['Data']
dfNew = df['Data'].str.split(' ', expand=True)
dfNew.to_csv('output.csv')
dfNew.shape
dfX = dfNew.iloc[:,2:]
dfY =dfNew.iloc[:,:2]
dfX=dfX.iloc[:,:-1]
dfX = dfX.astype(float)
dfY = dfY.astype(float)
```

Made this simple neural network architecture with last layer having 2 perceptrons (for two different labels) and since this is a regression problem the output layer has no activation function

```python
model=Sequential()
model.add(Dense(64,activation='relu',input_dim=39))
model.add(Dense(32,activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(2))
model.compile(loss='mean_squared_error', optimizer='adam',metrics=['mae'])
```

```python
model.fit(dfX,dfY,epochs=100,batch_size=50)
```

Tested the data on test.csv

```python
dfTest = pd.read_csv('/content/test.csv',header=None)
dfTest.columns=['Data2']
dfTestNew = dfTest['Data2'].str.split(' ', expand=True)
dfTestNew.shape
dfTestX = dfTestNew.iloc[:,2:]
dfTestY =dfTestNew.iloc[:,:2]
dfTestX=dfTestX.iloc[:,:-1]
dfTestX = dfTestX.astype(float)
dfTestY = dfTestY.astype(float)
```

```python
yPred = model.predict(dfTestX)
yPred
```

found mean absolute error and R2 score for the test data

```python
m =mean_absolute_error(yPred,dfTestY)
m
dfYpred = pd.DataFrame(yPred)
```

```python
r2 = r2_score(dfTestY, dfYpred)
print(f"R² Score: {r2}")
```
